# Medidas de grafo 

Primero generamos un grafo para los ejemplos: 

```python
import networkx as nx
import matplotlib.pyplot as plt
import scipy as sp

G=nx.Graph()

nodos=(1,2,3,4)
edges=[(1,2),(2,3),(3,1),(4,2)]


G.add_nodes_from(nodos)
G.add_edges_from(edges)  

G[1][3]['color'] = "blue"

nx.draw(G,with_labels=True)
plt.show()
```
